## PCA and Dimensionality Reductions

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
```

```python
standard_scaler = StandardScaler()
```

```python
standardized_data = standard_scaler.fit_transform(filtered_df_3.iloc[:, 2:])
```

```python
standard_df = pd.DataFrame(standardized_data,
                           columns = filtered_df_3.iloc[:,2:].columns)
```

```python
standard_df
```

```python
pipeline = Pipeline(
    [
        ('scaler', StandardScaler()),
        ('model', PCA(n_components=4))
    ]
)
```

```python
pipeline.fit(filtered_df_3.iloc[:,2:])
```

```python
new_data = pipeline.fit_transform(filtered_df_3.iloc[:,2:])
```

```python
new_df = pd.DataFrame(new_data,
                     columns = [f'pc_{i}' for i in range(4)])
```

```python
new_df
```

```python
pca_model = pipeline.named_steps["model"]
```

```python
components = pca_model.components_
```

```python
components_df = pd.DataFrame(components,
                            columns = filtered_df_3.iloc[:, 2:].columns,
                            index=[f'pc_{i}' for i in range(4)])
```
